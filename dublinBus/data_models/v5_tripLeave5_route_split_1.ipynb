```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
# Read tripleave5 CSV file into Data Frame:
wk = pd.read_csv('v5_tripLeave5_route_split_1.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
wk.shape
```

```python
wk.head()
```

```python
wk.dtypes
```

## Datatype Changes

```python
wk['num_lineID'] = wk['num_lineID'].astype('int16')
wk['DIRECTION'] = wk['DIRECTION'].astype('int16')
```

```python
wk['actual_duration'] = wk['actual_duration'].astype('int64')
wk['dayOfWeek'] = wk['dayOfWeek'].astype('int16')
```

```python
wk['weekend'] = wk['weekend'].astype('int16')
wk['rushHour'] = wk['rushHour'].astype('int16')
```

```python
wk.head()
```

```python
wk.dtypes
```

## Route 151

```python
lwk = wk[(wk['LINEID'] == '151')]
```

```python
lwk.shape
```

```python
# Drop Route 151 rows from wk
wk.drop(wk[wk['LINEID'] == '151'].index, inplace = True)
```
