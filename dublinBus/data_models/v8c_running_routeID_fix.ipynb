```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
rt = pd.read_csv('v7_route_9.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
rt.shape
```

```python
rt.head()
```

```python
rt = rt.drop(['PLANNEDTIME_DEP', 'ACTUALTIME_ARR', 'ACTUALTIME_DEP'], axis=1)
```

```python
rt.head()
```

## Create ID

```python
rt['str_dos'] = rt['DAYOFSERVICE'].astype('string')
```

```python
rt['str_trp'] = rt['TRIPID'].astype('string')
```

```python
rt['str_prog'] = rt['PROGRNUMBER'].astype('string')
```

```python
rt['str_stp'] = rt['STOPPOINTID'].astype('string')
```

```python
rt['dos_trpID_stpID_prgNum'] = rt['str_dos'] + '-' + rt['str_trp'] + '-' + rt['str_stp'] + '-' + rt['str_prog']
```

```python
rt.head()
```

## Drop Columns

```python
rt = rt.drop(['str_dos', 'str_trp', 'str_stp', 'str_prog'], axis=1)
```

```python
rt.head()
```

## Send to CSV

```python
# Send route to csv:
rt.to_csv('v8c_route_9.csv', index=False)
```
