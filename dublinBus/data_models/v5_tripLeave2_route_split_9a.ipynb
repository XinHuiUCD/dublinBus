```python
# Library Imports.
import pandas as pd

import numpy as np
```

```python
# Read tripleave2 route split 9a CSV file into Data Frame:
wk = pd.read_csv('v5_tripLeave2_route_split_9a.csv', keep_default_na=True, delimiter=',', skipinitialspace=True)
```

```python
wk.shape
```

```python
wk.head()
```

```python
wk.dtypes
```

## Datatype Changes

```python
wk['num_lineID'] = wk['num_lineID'].astype('int16')
wk['DIRECTION'] = wk['DIRECTION'].astype('int16')
```

```python
wk['actual_duration'] = wk['actual_duration'].astype('int64')
wk['dayOfWeek'] = wk['dayOfWeek'].astype('int16')
```

```python
wk['weekend'] = wk['weekend'].astype('int16')
wk['rushHour'] = wk['rushHour'].astype('int16')
```

```python
wk.head()
```

```python
wk.dtypes
```

## Route 51X

```python
lwk = wk[(wk['LINEID'] == '51X')]
```

```python
lwk.shape
```

```python
# Drop Route 51X rows from wk
wk.drop(wk[wk['LINEID'] == '51X'].index, inplace = True)
```

```python
wk.head()
```

```python
wk.shape
```
